```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("data/linear.csv")
df.head()
```

```python
df1 = df[df.tipe == "data1"]
plt.scatter(x = df1.X, y = df1.y, label="data1")
```

```python
df2 = df[df.tipe == "data2"]
plt.scatter(x = df2.X, y = df2.y, label="data2")
```

```python
df1 = df[df.tipe == "data1"]
df2 = df[df.tipe == "data2"]
plt.scatter(x = df1.X, y = df1.y, label="data1")
plt.scatter(x = df2.X, y = df2.y, label="data2")

x = [4,8,12,16,1,4,9,16]
y = [1,4,9,16,4,8,12,3]
label = [0,1,2,3,0,1,2,3]
colors = ['red','green','blue','purple']

fig = plt.figure(figsize=(8,8))
plt.scatter(x, y, c=label, cmap=matplotlib.colors.ListedColormap(colors))

cb = plt.colorbar()
loc = np.arange(0,max(label),max(label)/float(len(colors)))
cb.set_ticks(loc)
cb.set_ticklabels(colors)
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x="X", y="y", data=df, hue="tipe", sizes=10)
sns.lineplot(x="y", y="X", data=df, hue="tipe", sizes=10)
sns.barplot(x="X", y="y", data=df)
sns.violinplot(x="X", y="y", data=df)
```

```python
plt.figure(figsize=(10,10))
sns.lmplot(x="X", y="y", data=df, hue="tipe")
```

```python
plt.figure(figsize=(10,10))
sns.distplot(df1.y, bins=20, label="data1")
```

### statistical summary

```python
df3 = pd.read_csv("data/cereal.csv", index_col="name")
df3.head()
```

### missing value

```python
plt.figure(figsize=(10,10))
sns.heatmap(df3.corr(), annot=True, cmap="coolwarm", square=True, fmt=".1f")
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df3.isna(), annot=True, cmap="Blues")
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(~(df3.isna()), cmap="Blues", cbar=False)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(~(df3.isna()), cmap="Blues")
```

### pairplot

```python
sns.pairplot(df3, kind="reg", diag_kind="kde", corner=True)
```

```python
sns.pairplot(df3, kind="reg", diag_kind="kde", corner=True, hue="shelf", vars=["calories", "fat", "rating"])
```

```python
sns.jointplot(x="potass", y="fiber", data=df3, kind="reg", marginal_kws={"bins": 25})
```
