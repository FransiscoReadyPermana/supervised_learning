```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/linear.csv")
df.head()
```

```python
df1 = df[df.tipe == "data1"]
plt.scatter(x = df1.X, y = df1.y, label="data1")
```

```python
df2 = df[df.tipe == "data2"]
plt.scatter(x = df2.X, y = df2.y, label="data2")
```

```python
df1 = df[df.tipe == "data1"]
df2 = df[df.tipe == "data2"]
plt.scatter(x = df1.X, y = df1.y, label="data1")
plt.scatter(x = df2.X, y = df2.y, label="data2")
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x="X", y="y", data=df, hue="tipe", sizes=10)
```

```python
plt.figure(figsize=(10,10))
sns.lmplot(x="X", y="y", data=df, hue="tipe")
```

```python
plt.figure(figsize=(10,10))
sns.distplot(df1.y, bins=20, label="data1")
```

### statistical summary

```python
df = pd.read_csv("data/cereal.csv", index_col="name")
df.head()
```

### missing value

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", square=True, fmt=".1f")
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.isna(), annot=True, cmap="Blues")
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(~(df.isna()), cmap="Blues", cbar=False)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(~(df.isna()), cmap="Blues")
```

### pairplot

```python
sns.pairplot(df, kind="reg", diag_kind="kde", corner=True)
```

```python
sns.pairplot(df, kind="reg", diag_kind="kde", corner=True, hue="shelf", vars=["calories", "fat", "rating"])
```

```python
sns.jointplot(x="potass", y="fiber", data=df, kind="reg", marginal_kws={"bins": 25})
```
