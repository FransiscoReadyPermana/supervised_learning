```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X = np.linspace(0, 10, 100)
y = np.sqrt(X)
y1 = np.sin(X)
y2 = np.cos(X)
```

### inisialisasi figure dan axes

```python
fig = plt.figure(figsize=(5, 5))
ax = plt.axes()
ax.plot(X, y, 'r-', label='y = sqrt(x)')
ax.plot(X, y1, 'b-', label='y = sin(x)')
```

```python
plt.figure(figsize=(5, 5))
plt.axes()
plt.plot(X, y, 'r-', label='y = sqrt(x)')
plt.plot(X, y1, 'b-', label='y = sin(x)')
```

```python
plt.figure(figsize=(5, 5))
plt.axes()
plt.plot(X, y2, 'rv', label='y = sqrt(x)',  markersize = 10)
plt.plot(X, y1, label='y = sin(x)', linewidth=30)
plt.xlim(0, 10)
plt.ylim(-2, 2)
plt.axis('equal')
plt.legend()
plt.title('Plotting sin(x) and cos(x)')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100) 

X, Y = np.meshgrid(x, y)
z = np.sqrt(X**2 + Y**2 + 10)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, z, rstride=1, cstride=1, cmap='viridis', edgecolor='none', antialiased = True)
ax.set_ylabel('Y')
ax.set_xlabel('X')
ax.view_init(elev=30, azim = 45)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, z, rstride=1, cstride=1, cmap='viridis', antialiased = True)
ax.set_ylabel('Y')
ax.set_xlabel('X')
ax.view_init(elev=30, azim = 45)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, z, 50, rstride=1, cstride=1, cmap='viridis', edgecolor='none', antialiased = True)
ax.set_ylabel('Y')
ax.set_xlabel('X')
ax.view_init(elev=30, azim = 45)
```
